```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("/content/popular-languages.csv")
df.head()
```

```python
df.drop(['Unnamed: 0'] , axis = 1 , inplace= True)
```

```python
df
```

```python
df.rename(columns = {'language name' : 'language' , 
                     'annual average salary' : 'salary'} ,inplace= True)
df
```

```python
df['salary']= df['salary'].str.replace(',' , '')
df['salary']= df['salary'].str.replace('$' , '')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['salary'] = pd.to_numeric(df['salary'])
```

```python
df.dtypes
```

```python
df.to_csv('popular-languages_edited.csv')
```
